```python
#%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
#print(city.head()) #To see similar columns
#print(ride.head())
tot = ride.merge(city,how="left",on="city",sort=False)
# Display the data table for preview
tot.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
cityGroups = tot.groupby(["type","city"])
xs = cityGroups["city"].count() #Number of rides, grouped by city, and those cities grouped by type
ys = cityGroups["fare"].mean() #Mean fare per city, with those cities grouped by type
rurX = xs.loc["Rural"] #These .loc all get the grouped cities and their associated values
rurY = ys.loc["Rural"]
urbX = xs.loc["Urban"]
urbY = ys.loc["Urban"]
subX = xs.loc["Suburban"]
subY = ys.loc["Suburban"]

sizes = cityGroups["driver_count"].mean()#Gets the driver count per city to use as bubble size.
rurS = sizes.loc["Rural"]*10#Adjusts the sizes to the right size for each type
urbS = sizes.loc["Urban"]*10
subS = sizes.loc["Suburban"]*10
# Build the scatter plots for each city types
plt.scatter(urbX,urbY,c="#f08080",alpha=.75,edgecolors="black",s=list(urbS),label="Urban")
plt.scatter(subX,subY,c="#87cefa",alpha=.75,edgecolors="black",s=list(subS),label="Suburban")
plt.scatter(rurX,rurY,c="gold",alpha=.75,edgecolors="black",s=list(rurS),label="Rural")
# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
# Create a legend
lgnd = plt.legend(loc="upper right",title="City Types")
lgnd.legendHandles[0]._sizes = [30]#Sets the size of the markers in the legend
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
# Incorporate a text label regarding circle size
plt.text(42,35,s="Note:\nCircle size correlates with driver count per city.")
# Save Figure
plt.savefig("BubblePlot.png",bbox_inches="tight")#bbox saves the text as well
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
cityFares = cityGroups["fare"].sum()
typeFares = cityFares.groupby(cityFares.index.get_level_values(0)).sum()#Regroups by type of city, and sums the city values
# Build Pie Chart
explode = [0,0,.1]
labels = ["Rural","Suburban","Urban"]
colors = ["gold","#87cefa","#f08080"]
plt.pie(typeFares, explode = explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True,startangle = 140)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("TotFaresPie.png",bbox_inches="tight")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
cityRides = cityGroups["city"].count()
typeRides = cityRides.groupby(cityRides.index.get_level_values(0)).sum()#Regroups by type of city, and sums the city values
# Build Pie Chart
explode = [0,0,.1]
labels = ["Rural","Suburban","Urban"]
colors = ["gold","#87cefa","#f08080"]
plt.pie(typeRides,explode=explode,labels=labels,colors=colors, autopct="%1.1f%%",shadow=True,startangle = 140)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("TotRidesPie.png",bbox_inches="tight")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
cityDrivers = cityGroups["driver_count"].mean()
typeDrivers = cityDrivers.groupby(cityDrivers.index.get_level_values(0)).sum()#Regroups by type of city, and sums the city values
# Build Pie Charts
explode = [0,0,.1]
labels = ["Rural","Suburban","Urban"]
colors = ["gold","#87cefa","#f08080"]
plt.pie(typeDrivers,explode = explode, labels = labels, colors = colors,autopct="%1.1f%%",shadow=True,startangle = 140)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("TotDrivesPie.png",bbox_inches="tight")
plt.show()
```
